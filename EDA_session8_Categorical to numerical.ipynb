**3) one hot encoder**
- one hot means if one is ON another one is OFF
- ON represents with 1, and OFF represents with 0
- For example case status has two uniqu labels
  - Denied
  - Certified
- If we apply one hot encoder on case status column it will create two extra columns
  - case_status_Denied
  - case_status_Certified

|case_status|case_status_denied|case_status_Certified|
|--------|--|--|
|Denied |1|0|
|Certified|0|1|


**Advantages**
- In machine learning we have one important property is **Independent of variavles**
- Machine learning models excepts variables are independent each other
- one hot encoder creating new columns, the new columns are **Independent each other**
- Independent variable means both variables have **90 degrees phase shift**
- 90 degreee phase shift also called as **perpendicular each other**
- Perpendicular means **Orthogonal each other**

**Disadvantage**
- one heat encoder creates new columns equal to number of unique labels
- If number of unique labels increase, new columns also increase
- The more columns leads data is more
- Then data is more the complexity will increase
- Complexity increase more computation power required
- Also time also increase
- The entire process names as **curse of dimentionality**
