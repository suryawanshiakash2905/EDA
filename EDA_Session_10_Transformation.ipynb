**Data Transformation**
- Generally data has 3 types of distributiona
  - Normal distributiona
  - Positive Skewed
  - Negative Skewed
- All the maths concepts developed by assumption as data follows Normal distributiona
- Data skew happens because of many resons, in that one reason is outlier
- once we perform the outlier analysis then we can apply transformation methods, if the datta does not follows normality
- we have some transformation methods mentioned below
  - Log Transformation
  - Exponential Transformation
  - Reciprocal Transformation
  - Square root Transformation
  - Power Transformation

**Log transformation**
- Log Transformation means performing logarithm operation on original data

```python
x=2
np.log(x)
```

```python
log_data=np.log(exp_data)
log_data
```

**power Transformation**
- Box cox
- ye-jhonson
- It is under sklearn preprocessing

```python
import numpy as np
from sklearn.preprocessing import PowerTransformer
power=PowerTransformer(method='yeo-johnson',standardize=True)
data_trans=power.fit_transform(exp_data.reshape(-1,1))
plt.hist(data_trans)
```

```python
import numpy as np
from sklearn.preprocessing import PowerTransformer
power=PowerTransformer(method='box-cox',standardize=True)
data_trans=power.fit_transform(exp_data.reshape(-1,1))
plt.hist(data_trans)
```
