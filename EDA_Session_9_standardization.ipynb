```python
visa_df=pd.read_csv(r"C:\Users\surya\NareshIT\ EDA_python\Visadataset - Visadataset.csv")
pv=visa_df['prevailing_wage']
mean=pv.mean()
sd=pv.std()
Nr=pv-mean
visa_df['pv_z']=Nr/sd
visa_df

```
