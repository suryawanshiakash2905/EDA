```python
unique_count=bank_df['month'].unique()
for i in unique_count:
    con=bank_df['month']==i
    values=len(bank_df[con])
    print(f"{i} : {values}")
    
```

```python
# use python to count the number of people in per month
mnt_count=bank_df['month'].unique()
count_list=[]
for i in mnt_count:
    con=bank_df['month']==i
    values=len(bank_df[con])
    count_list.append(values)
month_df=pd.DataFrame(zip(mnt_count,count_list),
                     columns=['Month','Count'])
month_df   
```

**Bar chart**

```python
plt.figure(figsize=(10,5)) #change the size of the figure
plt.bar('Month','Count',data=Month_df) #plot
plt.title('Bar Chart') #Title of the plot
plt.xlabel('Month') # x-axis name
plt.ylabel('Count') # y-axis name
plt.show()
```

```python
plt.figure(figsize=(10,5))
mnc=bank_df['month'].value_counts()
keys=mnc.keys()
sns.countplot(data=bank_df,x='month',order=keys)
plt.title('Bar chart')
plt.show()
```
